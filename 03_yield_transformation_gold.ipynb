```python
df = spark.read.format("delta").load("dbfs:/crop/silver/crop_clean")
```

```python
# Yield = Production / Area

from pyspark.sql.functions import col

df = df.withColumn(
    "Yield",
    col("Production") / col("Area")
)
```

```python
# handle division errors
df = df.fillna({"Yield":0})
```

```python
# Gold Layer
df.write \
    .format("delta") \
    .mode("overwrite") \
    .save("dbfs:/crop/gold/crop_yield")
```
